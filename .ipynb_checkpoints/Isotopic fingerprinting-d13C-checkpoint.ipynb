```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
%matplotlib inline
```

```python
# Import data
df = pd.read_csv('~/Documents/GitHub/data/schiff c-csiaa data from schiff_larsen_in prog.csv')
df = df[['Group ID','Group ID2','Feature', 'Ala', 'Asx', 'Glx', 'Gly', 'Ser', 'Pro', 'Ile', 'Leu', 'Lys', 'Phe', 'Thr', 'Tyr', 'Val']]
df
```

```python
# Isolate EAA
df = df[['Group ID','Feature', 'Phe', 'Thr', 'Ile', 'Leu', 'Val']]
df
```

```python
# Split into two datasets
df_train = df[(df['Feature'] == 'Cyanobacteria') | (df['Feature'] == 'Macroalgae') 
              | (df['Feature'] == 'Microalgae') | (df['Feature'] == 'Bacteria')]
df_train = df_train[['Group ID','Feature', 'Phe', 'Thr', 'Ile', 'Leu', 'Val']]
df_train = df_train.dropna()

leio = df[(df['Feature'] == 'Leiopathes-post') | (df['Feature'] == 'Leiopathes-pre') |
         (df['Feature'] == 'Sediment trap')]
df_test = leio
```

```python
df_train
```

```python
df_test
```

```python
df_pca = pd.concat([df_train, df_test], axis=0)
df_pca = df_pca.reset_index()

groups = ['Cyanothece (NO3)', 'C1', 'C2']
df_pca = df_pca[~df_pca['Group ID'].isin(groups)] # Drop unwanted groups

df_pca = df_pca.reset_index()
df_pca
```

```python
# PCA
# Separating out the features
features = ['Phe', 'Thr', 'Ile', 'Leu', 'Val']

x = df_pca.loc[:, features].values
xmean = x.mean(axis=1, keepdims = True)
x = x - xmean # Normalize to the mean of each row

# Separating out the target
y = df_pca.loc[:,['Feature']].values
# Standardizing the features
x = StandardScaler(with_std=True, with_mean=True).fit_transform(x)

pca = PCA(n_components=4)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents, 
                           columns = ['principal component 1', 'principal component 2', 
                                      'principal component 3', 'principal component 4'])
```

```python
# principalDf
```

```python
finalDf = pd.concat([principalDf, df_pca[['Feature']]], axis = 1)
# print(finalDf)

new_array = finalDf['Feature'].unique()
new_list = np.ndarray.tolist(new_array)
new_list
```

```python
# 2D visualization
# Visualize 2D
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15, fontname = 'Helvetica')
ax.set_ylabel('Principal Component 2', fontsize = 15, fontname = 'Helvetica')
# ax.set_title('PCA', fontsize = 15, fontname = 'Helvetica')
ax.set_xlim(-4,6)
ax.set_ylim(-4,4)
targets = new_list

colors = ["#999999", "#E69F00", "#56B4E9", "#009E73", "#F0E442", "#0072B2", "#D55E00", "#CC79A7"]
markers = ['o', 'o','v', 's', 'v', 'D', 'D', '*']

for tick in ax.get_xticklabels():
    tick.set_fontname("Helvetica")
    tick.set_fontsize(14)
for tick in ax.get_yticklabels():
    tick.set_fontname("Helvetica")
    tick.set_fontsize(14)
for target, color, marker in zip(targets,colors,markers):
    indicesToKeep = finalDf['Feature'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 150,
              linewidths=0.75,
              edgecolors = 'k',
              linestyle = 'None',
              marker = marker)
    plt.grid('None')
ax.legend(targets, frameon = False, markerscale = 0.75)
ax.vlines(x=0,ymin=-10,ymax=10,linestyle='dashed',linewidth=1,label=None)
ax.hlines(y=0,xmin=-10,xmax=10,linestyle='dashed',linewidth=1,label=None)
ax.grid()

# Save figure
plt.savefig('pca_d13c.png', dpi=300, bbox_inches='tight')
```
