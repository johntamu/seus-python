## Chapter 1 data analysis and figures

```python
# Load packages
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
pd.set_option('max_rows', 6)
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
import seaborn as sns
import cartopy
import cartopy.feature as cfeature
import cmocean.cm as cmo
from matplotlib.ticker import LogFormatter 
import matplotlib.ticker as mticker
import csv
```

```python
# Import data
path1 = '~/Documents/GitHub/data/schiff_bulk_years_09-04-2019.csv'

df = pd.read_csv(path1, comment = '#')
df.rename(columns={"d15n.vs.air": "d15n", "d13c.vs.vpdb": "d13c", "linear.ad" : "yrAD", "coral.id" : "coral"}, inplace = True)
df
```

```python
x = "Year CE"
phe = r'$\mathrm{\delta}^{15}$N Phe ('u"\u2030)"
eaa_neaa = ['Phe', 'Thr', 'Ile', 'Leu', 'Val', 'Asx', 'Glx', 'Pro', 'Ala', 'Ser', 'Gly'] # For Essential/Non-Essential ordering
tr_srcaa = ['Glu', 'Asp', 'Ala', 'Ile', 'Leu', 'Pro', 'Val', 'Gly', 'Ser', 'Lys', 'Tyr', 'Phe', 'Thr'] # For Trophic/Source AA ordering
n = r'$\mathrm{\delta}^{15}$N ('u"\u2030)"
c = r'$\mathrm{\delta}^{13}$C ('u"\u2030)"
```

```python
# Split the dataframe by coral
df_stet = df[df.coral == 'stet-4904-bc1-d2']
df_jack4907 = df[df.coral == 'jack-4907-bc1-d3']
df_sav = df[df.coral == 'sav-4902-bc1-unk']
df_jack4684 = df[df.coral == 'jack-4684-bc-unk']
```

### Bulk d15N

```python
# Recent history, bulk d15N
fig = plt.figure()
ax = fig.add_subplot(111)

plt.plot('yrAD', 'd15n', data=df_stet,
         color='#e66101', label = 'Stetson Banks', markeredgecolor = 'black', markersize = 8)
plt.plot('yrAD', 'd15n', data=df_jack4684, 
         color='#5e3c99', label = 'Stetson Banks', markeredgecolor = 'black', markersize = 8)

# Rolling mean



# Draw x label
ax.xaxis.set_label_position('top') # this moves the label to the top
ax.xaxis.set_ticks_position('top') # this moves the ticks to the top
ax.set_xlabel("Year CE")

# Draw y label
ax.set_ylim(ax.get_ylim())
ax.set_ylabel(n)

```

```python
# Ancient d15N changes

fig = plt.figure()
ax = fig.add_subplot(111)

plt.plot('yrAD', 'd15n', data=df_sav,
         color='#a6611a', label = 'Stetson Banks', markeredgecolor = 'black', markersize = 8)
plt.plot('yrAD', 'd15n', data=df_jack4907, 
         color='#018571', label = 'Stetson Banks', markeredgecolor = 'black', markersize = 8)

# Rolling mean



# Draw x label
ax.xaxis.set_label_position('top') # this moves the label to the top
ax.xaxis.set_ticks_position('top') # this moves the ticks to the top
ax.set_xlabel("Year CE")

# Draw y label
ax.set_ylim(ax.get_ylim())
ax.set_ylabel(n)
```
